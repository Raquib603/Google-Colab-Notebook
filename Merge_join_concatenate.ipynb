```python
import pandas as pd
```

```python
df1 = pd.read_csv('/content/LOTR.csv')
```

```python
df2 = pd.read_csv('/content/LOTR 2.csv')
```

```python
df1
```

```python
df2
```

```python
#default merge will inner join
df1.merge(df2, how = 'inner')
```

```python
#outter join
df1.merge(df2, how = 'outer')
```

```python
#everything from the left(df1) and add
df1.merge(df2, how = 'left')
```

```python
#everythin from the left(df2) and add similar from (df1)
df1.merge(df2, how ='right')
```

```python
df1.merge(df2, how = 'cross')
```

#Join Function


```python
df1.join(df2, on = 'FellowshipID', how ='outer', lsuffix='_left', rsuffix='_right')
```

```python
#just add on one another
pd.concat([df1, df2])
```

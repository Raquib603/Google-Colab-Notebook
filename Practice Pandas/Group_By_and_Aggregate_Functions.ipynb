```python
import pandas as pd
```

```python
df = pd.read_csv('/content/Flavors.csv')
df
```

# Group by
we need something that has similar multiple values like vanilla
and chocolate in this case

```python
df.set_index('Base Flavor', inplace =True)
df
```

```python
df.sort_index(inplace=True)
df
```

```python
frame_group_by = df.groupby('Base Flavor')
```

```python
# Calculate the mean of only numerical columns and showing the mean of the data
frame_group_by.mean(numeric_only=True)
```

```python
#Group and count the numbers
df.groupby('Base Flavor').count()
```

```python
df.groupby('Base Flavor').sum()
```

```python
df.groupby('Base Flavor').sum(numeric_only=True)
```

#Aggregate Functions
Now we can group by a column and
see multiple function on one column together

```python
df.groupby('Base Flavor').agg({'Flavor Rating':['mean','max','count','sum']})
```

# Shortcut

```python
df.groupby('Base Flavor').describe()
```

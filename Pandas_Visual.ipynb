```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Ice Cream Ratings.csv')
df
```

```python
df.columns
```

```python
print(plt.style.available)
```

```python
#use any of the style
plt.style.use('_classic_test_patch')
```

```python
#default is line plot
df.plot()
```

```python
#subplots for seperating the plots
df.plot(kind = 'line', subplots = True)
```

#Extra things

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Rating', ylabel = 'Scores' )
```

#Bar Graph

```python
df.plot(kind = 'bar', stacked = True, title = 'Ice Cream Rating', xlabel = 'Daily Rating', ylabel = 'Scores')
```

#Work with only one column on bar chart

```python
#One column with bar grapf
df['Flavor Rating'].plot(kind = 'bar', title = 'Rating')
```

```python
#Bar graph horizontally
df.plot.barh(x = 'Flavor Rating', title = 'Rating')
```

#Pie Char

```python
df.plot.pie(y = 'Flavor Rating', figsize = (10, 6))
```

#Scatte Plot

```python
#by taking two columns
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 100, c = 'Red')
```

#Histogram

```python
df.plot.hist()
```

```python
df.plot.hist(bins = 100)
```

#Box Plot

```python
df.boxplot()
```

```python
df.plot.area(figsize = (10, 5))
```
